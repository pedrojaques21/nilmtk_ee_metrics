```python
import re
import compress_pickle as cp
import pandas as pd
```

```python
ee_metrics_dic = cp.load('ee_metrics_dic.gz')
ee_metrics_dic_cep = cp.load('ee_metrics_dic_cep.gz')
ee_metrics_dic_teca = cp.load('ee_metrics_dic_teca.gz')
ee_metrics_dic_dev = cp.load('ee_metrics_dic_dev.gz')
ee_metrics_dic_MRuMR_macro = cp.load('ee_metrics_dic_MRuMR_macro.gz')
```

```python
ee_metrics_x_experiments = {}
```

```python
for metric in ee_metrics_dic:
    if re.search('DW.gz$', metric):
        ee_metrics_x_experiments.update({metric : ee_metrics_dic[metric]['dish washer']})
    elif re.search('FF.gz$', metric):
        ee_metrics_x_experiments.update({metric : ee_metrics_dic[metric]['fridge freezer']})
    elif re.search('FR.gz$', metric) or re.search('FRI.gz$', metric):
        ee_metrics_x_experiments.update({metric : ee_metrics_dic[metric]['fridge']})
    elif re.search('FZ.gz$', metric):
        ee_metrics_x_experiments.update({metric : ee_metrics_dic[metric]['freezer']})
    elif re.search('KT.gz$', metric):
        ee_metrics_x_experiments.update({metric : ee_metrics_dic[metric]['kettle']})
    elif re.search('MW.gz$', metric):
        ee_metrics_x_experiments.update({metric : ee_metrics_dic[metric]['microwave']})
    elif re.search('TV.gz$', metric):
        ee_metrics_x_experiments.update({metric : ee_metrics_dic[metric]['television']})
    elif re.search('WM.gz$', metric):
        ee_metrics_x_experiments.update({metric : ee_metrics_dic[metric]['washing machine']})
```

```python
data = cp.load('ee_metrics_x_experiments.gz')
```
